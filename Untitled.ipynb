```python
# resume_screening_tool.ipynb

import pandas as pd
import re
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load the dataset
df = pd.read_csv("UpdatedResumeDataSet.csv")

# Basic info
print(df.head())
print(df['Category'].value_counts())

# Preprocess resume text
def clean_text(text):
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z ]", "", text)
    text = text.lower()
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

df['Cleaned_Resume'] = df['Resume'].apply(clean_text)

# Generate word cloud
def generate_wordcloud(text):
    wc = WordCloud(width=800, height=400, background_color='white').generate(' '.join(text))
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()

generate_wordcloud(df['Cleaned_Resume'])

# Keyword Screening Function
def keyword_match(resume_text, keywords):
    matches = []
    for keyword in keywords:
        if re.search(r'\b' + re.escape(keyword.lower()) + r'\b', resume_text.lower()):
            matches.append(keyword)
    return matches

# Define keywords
keywords = ['python', 'data analysis', 'machine learning', 'sql', 'excel']

# Apply keyword matching
df['Matched_Keywords'] = df['Cleaned_Resume'].apply(lambda x: keyword_match(x, keywords))

# Print top 10 resumes with matching keywords
print(df[['Category', 'Matched_Keywords']].head(10))
```
